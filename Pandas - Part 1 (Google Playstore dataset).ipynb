# Slicing through loc and iloc

1. iloc will use when we have integer indexing (consider the excel default row no.s on left side)
2. loc will use when we have label indexing    (consider the serial numbers that we add in excel)

    

<img src = "https://www.exceldemy.com/wp-content/uploads/2022/07/subtotal-formula-in-excel-for-serial-number-15.png">





suppose if we delete the record of Keat the serial number will be changed to 1,2,4,5.....but
the excel default line numbers wont be changed . so here serial no provided by user is label indexing
while default row number of excel is integer indexing

```python
playstore.iloc[0]                                # to view the details of first record using integer indexing
```

```python
playstore.iloc[0,-4]                # to view the value available in 4th last column (genres) of 1st row using integer indexing
```

```python
playstore.iloc[5:11]                # to view the data of records from 5th record to 10th record using int indexing
```

```python
playstore.iloc[5:11, :4]       # to view the first three columns data for records from 5th to 10th using integer indexing
```

```python
playstore.iloc[[5,8], :2]       # to view the first three columns data for record# 5 and 8 using integer indexing
```

## Logical Operator

* `&` and
* `|` or
* `~` not

```python
playstore["Category"].unique()       # distinct values in Category column
```

```python
playstore[(playstore["Category"] == "GAME") & (playstore["Rating"] > 4.0 )]   # 2 criteria where both must be true
```

```python
playstore[(playstore["Category"] == "GAME") & ~(playstore["Type"] == "Paid")]       # 2 criteria where first must be true 
                                                                                    # while seconds mustn't be
```

```python
playstore.info()                 # Since the dtype of Reviews is object, we need to change it to int
```

```python
playstore['Reviews'] = pd.to_numeric(playstore['Reviews'])           # to convert dtype of Reviews column to int
```

```python
playstore.iloc[10472,3]             # after getting error, we need to do cleaning on 'position 10472'
```

```python
playstore['Reviews'] = pd.to_numeric(playstore['Reviews'])      # to convert dtype of Reviews column to int now after cleaning
```

```python
playstore.loc[(playstore["Reviews"] >= 1000) & ~(playstore["Type"] == "Free"), ["App","Category","Price","Installs","Size",'Reviews']]
# boolean indexing as above with multi columns display
```
